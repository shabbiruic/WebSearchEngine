```python
from tkinter import *
import tkinter
from Search import search 
import webbrowser
from PIL import ImageTk, Image

url_count = 50

# creating the Window on which various UI Components will lay. 
window = Tk()
window.title("Search Engine")
window.geometry('600x200')

# function to make each entry of list clickable and open the browser new tab.
def weblink(*args):
    index = listbox.curselection()[0]
    item = listbox.get(index)
    webbrowser.open_new(item)

# this is the list box where all current batch of 10 URL will be displayed
listbox = Listbox(window,width=80)
listbox.grid(column=0,row=1)

# text box where user will enter the query
textbox = Entry(window,width=50)
textbox.grid(column=0,row=0)

# creating global parameters for keeping track of index of URL List
start=0
end=10

# Global list which will hold the fetched urls from search engine 
list=[]

# Gets called when more links button is pressed
def more():
    previousBtn['state']= 'normal'
    global start
    global end
    listbox.delete(0, END)
    start = end
    end = end + 10
    for item in list[start:end]:
        listbox.insert(END, item)
    if(end >= url_count):
        moreBtn['state']='disable'
        return

# gets called when previous links button is pressed 
def previous():
    global start
    global end
    moreBtn['state']='normal'
    listbox.delete(0, END)
    start = start - 10
    end = end - 10
    for item in list[start:end]:
        listbox.insert(END, item)
    if(start <= 0):
        previousBtn['state']='disable'
        return

#     gets called when clear button is pressed.
def clear():
    global start
    global end
    global list
    start=0
    end=10
    list=[]
    moreBtn['state']='disable'
    previousBtn['state']= 'disable'
    clearBtn['state']='disable'
    listbox.delete(0, END)
    textbox.delete(0, END)

#     gets Called when search button is pressed.
def onSearch():
    global start
    global end
    global list
    start =0
    end = 10
    listbox.bind('<<ListboxSelect>>', weblink)
    moreBtn['state']='normal'
    previousBtn['state']= 'disable'
    clearBtn['state']='normal'
    listbox.delete(0, END)
    query = textbox.get()
    list = search(query,url_count)
    for item in list[start:end]:
        listbox.insert(END, item)
        
# Creating more button which will fetch the further links
moreBtn = Button(window, text="More Links",command=more)
moreBtn.grid(column=2,row=1)
moreBtn['state']= 'disable'

# Creating previous button which will show the already seen links again 
previousBtn = Button(window, text="Previous Links",command=previous)
previousBtn.grid(column=1,row=1)
previousBtn['state']= 'disable'

# will fetch the new URL links depending on query present in text box
searchBtn = Button(window, text="Search",command=onSearch)
searchBtn.grid(column=1,row=0)

# This will clear the previous search results and query.
clearBtn = Button(window, text="Clear",command=clear)
clearBtn.grid(column=2,row=0)
clearBtn['state']= 'disable'

# showing the window untill user closes it
window.mainloop()
```
